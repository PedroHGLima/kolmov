```python
import os
from kolmov.core.base_exporter import export_fastnet_to_onnx
```

```python
print('Get tunings from: \n %s' %os.environ['OLD_TUNINGS'])
```

```python
wanted_tuning = 'final_tuning_09032019_athena'
tuning_path   = os.path.join(os.environ['OLD_TUNINGS'], wanted_tuning)
print('Using tunings in: \n %s' %(tuning_path))
# configure the file
m_version            = 'v1'
m_model_format_name  = 'data17_13TeV_EGAM2_probes_lhmedium_EGAM7_vetolhvloose.model_v1.electron%s.et%d_eta%d.onnx'
m_output_conf_name   = 'ElectronRinger%sTriggerConfig.conf'
m_tuning_name        = 'TrigL2_20190319_v1_athena_legacy'
export_path          = tuning_path+'_onnx'
print('Using this path to export: \n %s' %export_path)
```

```python
base_export = export_fastnet_to_onnx(tuning_path)
for ioperation in ['Tight', 'Medium', 'Loose', 'VeryLoose']:
    print('Translate and exporting tuning in %s' %(ioperation))
    base_export.create_config_files(operation_point=ioperation, 
                                    tuning_name=m_tuning_name,
                                    version=m_version,
                                    model_format_name=os.path.join(export_path+'/models', m_model_format_name),
                                    output_config_name=os.path.join(export_path, m_output_conf_name))
```
